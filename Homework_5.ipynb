```python
#import necessary packages
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#define variables
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, '-b', linestyle="--") #create a blue plot with dashed lines
plt.title('Basic Plotting with Matplotlib') #add a title
plt.xlabel('X') #add an x-axis label
plt.ylabel('sin(X)') #add a y-axis label
plt.show() #show the plot
```

```python
#load the dataset
tips = sns.load_dataset("tips")

#create a histogram
plt = sns.histplot(data = tips,
             x = "total_bill", #choose the x variable
             bins = 20, #set the number of bins to 20
             kde = True) #overlay a kde plot

plt.set_title("Distribution of Total Bill") #set the title
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#define a figure with two subplots
fig = plt.figure()
fig, axes = plt.subplots(1, 2)

#create the boxplot
sns.boxplot(x="time", y="tip", data=tips, ax=axes[0]) #assign the boxplot to axes[0]
axes[0].set_title("Tips Received by Time of Day") #set the title
axes[0].set_xlabel("Time of Day") #set the x label
axes[0].set_ylabel("Tips Received") #set the y label

#create the violin plot
sns.violinplot(x="time", y="tip", data=tips, ax=axes[1]) #assign the violin plot to axes[1]
axes[1].set_title("Tips Received by Time of Day") #set the title
axes[1].set_xlabel("Time of Day") #set the x label
axes[1].set_ylabel("Tips Recieved") #set the y label
```

```python
#create a scatterplot with a regression line for each group
sns.lmplot(tips,
           x = "total_bill",
           y = "tip",
           hue = "smoker", #color by smoker status
           legend = True) #include a legend
```

```python
#create the correlation matrix using numerical variables
correlation_matrix = tips.select_dtypes(include=['number']).corr()

#create a heatmap
sns.heatmap(correlation_matrix, #default is diverging color palette
            annot=True, #add correlation coefficient to each cell
            vmin=-1, #set color scale
            vmax=1,
            center=0)

#add a title and show the plot
plt.title("Correlation Matrix of Tips Dataset")
plt.show()
```

```python
#create a pairplot graph
g = sns.pairplot(data=tips,
             hue="time") #color the plots by time

g.map_diag(sns.kdeplot) #add kde plots along the diagonal
```
